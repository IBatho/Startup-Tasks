```python
from dataclasses import dataclass
from typing import List

@dataclass
class Order:
    order_id: int
    arrival_time: float
    deadline: float
    route: List[str]
    priority_level: int




```

```python
def parse_route(route_str: str) -> List[str]:
    return route_str.split("->")

if parse_route("A->B->D") == ['A', 'B', 'D'] and parse_route("B->C") == ['B', 'C']:
    print("✅ parse_route works!")
```

```python
import csv
from datetime import datetime

def load_orders(filename: str) -> List[Order]:
    orders: List[Order] = []
    with open(filename, newline = '') as f:
        reader = csv.DictReader(f)
        for row in reader:
            order = Order(
                order_id = int(row['order_id']),
                arrival_time = datetime.strptime(row['arrival_time'], "%Y/%m/%d/%H:%M:%S").timestamp(),
                deadline = float(row['deadline']),
                route = parse_route(row['route']),
                priority_level = int(row['priority_level'])
            )
            orders.append(order)
            
        
    return orders
```

```python
orders = load_orders('sample_order.csv')
print(orders)
```

```python
for o in orders:
    print(o.order_id, o.arrival_time, o.deadline, o.deadline > o.arrival_time)
```
